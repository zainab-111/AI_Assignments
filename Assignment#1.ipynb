# First Step, Plotting a Point

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating a numpy array
X = np.array([1])
Y = np.array([5])

# Plotting point using sactter method
plt.scatter(X,Y)
plt.show()
```

# Second Step, Plotting a line on grid 

```python
import numpy as np 
from matplotlib import pyplot as plt 

x = np.arange(1,30) 
y = 2*x

plt.title("Matplotlib demo") 
plt.xlabel("x-axis") 
plt.ylabel("y-axis")

plt.plot(x,y) 
plt.grid(b=True)
plt.show()
```
